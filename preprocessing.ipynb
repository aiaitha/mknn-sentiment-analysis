```python
import pandas as pd
```

```python
#load dataset
def load_data():
    # data = pd.read_csv('shopeetambahan.csv', sep=';')
    data = pd.read_excel('Dataset.xlsx')
    pd.set_option('display.max_colwidth', None)
    return data
data = load_data()
selected_columns = ['rating', 'reviews']
selected_data = data[selected_columns]
data = selected_data
data
```

```python
def remove(text):
    # Hapus angka
    text = ''.join([char if not char.isdigit() else ' ' for char in text])
    # Hapus tanda baca (selain huruf dan spasi)
    text = ''.join([char if char.isalnum() or char.isspace() else ' ' for char in text])
    # Hapus karakter non-ASCII (termasuk emoji) secara manual
    text = ''.join([char if ord(char) < 128 else '' for char in text])
    # Hapus spasi berlebih
    text = ' '.join(text.split())
    return text
data['reviews'] = data['reviews'].fillna('').astype(str)
data['cleaning'] = data['reviews'].apply(remove)
data
```

```python
def case_folding(text):
    # Ubah semua karakter menjadi huruf kecil
    result = ''
    for char in text:
        if 'A' <= char <= 'Z':
            result += chr(ord(char) + 32)
        else:
            result += char
    return result
data['case_folding'] = data['cleaning'].apply(case_folding)
data
```

```python
def tokenize(sentence):
    # Membagi kalimat menjadi kata-kata berdasarkan spasi
    words = []  # List untuk menyimpan kata-kata
    word = ''  # Variabel untuk menampung karakter-karakter yang membentuk kata
    for char in sentence:
        if char != ' ':  # Jika karakter bukan spasi
            word += char  # Tambahkan karakter ke dalam kata
        else:
            if word:  # Jika ada kata yang terakumulasi
                words.append(word)  # Tambahkan kata ke list
                word = ''  # Reset untuk kata berikutnya
    if word:  # Jika masih ada kata yang terakumulasi setelah loop selesai
        words.append(word)  # Tambahkan kata terakhir ke list
    return words

data['tokenization'] = data['case_folding'].apply(tokenize)
data
```

```python
import pandas as pd

# Fungsi normalisasi yang mencakup pemuatan kamus dan penerapan normalisasi pada kalimat
def normalisasi(sentence): 
    # Memuat kata-kata normalisasi dari file Excel
    normalizad_word = pd.read_excel("normalisasi.xlsx")  # Pastikan path benar
    normalizad_word_dict = {}  # Kamus untuk menyimpan kata normalisasi
    for index, row in normalizad_word.iterrows():
        normalizad_word_dict[row[0]] = row[1]  # Asumsi kolom pertama adalah kata, kolom kedua adalah pengganti

    # Iterasi setiap kata di dalam list token dan normalisasi
    normalized_sentence = []
    for word in sentence:
        # Langsung cek di kamus normalisasi
        normalized_word = normalizad_word_dict.get(word, word)  # Jika tidak ditemukan, gunakan kata aslinya
        normalized_sentence.append(normalized_word)  # Tambahkan kata yang sudah dinormalisasi

    return normalized_sentence  # Mengembalikan sebagai list token

# Menerapkan normalisasi pada tokenisasi di DataFrame
data["normalization"] = data["tokenization"].apply(normalisasi)
data
```

```python
# Membaca file stopwords.xlsx
stopwords_df = pd.read_excel('stopwords.xlsx')

# Ambil kolom 'stopwords' langsung, asumsikan kolom ini selalu ada
custom_stopwords = set(stopwords_df['stopwords'].dropna())  # Ganti 'stopwords' dengan nama kolom yang sesuai

# Fungsi untuk menghapus stopwords
def remove_stopwords(tokens):
    # Filter token yang bukan stopwords
    return [token for token in tokens if token not in custom_stopwords]

# Terapkan fungsi pada kolom 'normalization'
data['stopwords'] = data['normalization'].apply(remove_stopwords)
data
```

```python
def load_kata_dasar(file_path):
    """
    Fungsi untuk memuat daftar kata dasar dari file teks.
    """
    with open(file_path, 'r', encoding='utf-8') as f:
        # Membaca seluruh file dan mengonversinya menjadi set kata dasar
        return set(f.read().splitlines())

def simple_stemming(word, kata_dasar):
    # Pastikan kata dalam bentuk string
    if not isinstance(word, str):
        return word  # Kembalikan jika bukan string

    # Cek apakah kata sudah menjadi bentuk dasar terlebih dahulu
    if word in kata_dasar:
        return word  # Jika sudah bentuk dasar, langsung kembalikan kata tersebut

    # Daftar imbuhan
    suffixes = ["lah", "kah", "ku", "mu", "nya", "kan", "an", "i"]
    prefixes = ["meng", "peng", "mem", "pem", "meny", "peny", "men", "pen", "ber", "ter", "se", "di", "ke", "per", "me", "pe"]

    # Penghapusan prefiks terlebih dahulu
    for prefix in prefixes:
        if word.startswith(prefix):
            # Mengubah prefiks sesuai aturan
            if prefix == "meng" and len(word) > 4 and word[4] in "aiueo":
                word = "k" + word[4:]
            elif prefix == "peng" and len(word) > 4 and word[4] in "aiueo":
                word = "k" + word[4:]
            elif prefix == "mem" and len(word) > 3 and word[3] in "aiueo":
                word = "p" + word[3:]
            elif prefix == "pem" and len(word) > 3 and word[3] in "aiueo":
                word = "p" + word[3:]
            elif prefix == "meny" and len(word) > 4 and word[4] in "aiueo":
                word = "s" + word[4:]
            elif prefix == "peny" and len(word) > 4 and word[4] in "aiueo":
                word = "s" + word[4:]
            elif prefix == "men" and len(word) > 3 and word[3] in "aiueo":
                word = "t" + word[3:]
            elif prefix == "pen" and len(word) > 3 and word[3] in "aiueo":
                word = "t" + word[3:]
            elif prefix == "pem" and len(word) > 3 and word[3] == "r":
                word = "p" + word[3:]
            else:
                word = word[len(prefix):]  # Hapus prefiks jika tidak ada aturan khusus
            break  # Setelah menghapus prefiks, berhenti mencari prefiks lainnya

    # Kembali ke pengecekan kata dasar setelah penghapusan prefiks
    if word in kata_dasar:
        return word

    # Penghapusan sufiks jika kata tidak dalam kata dasar
    for suffix in suffixes:
        if word.endswith(suffix):
            word = word[:-len(suffix)]
            # Cek setelah menghapus sufiks
            if word in kata_dasar:
                return word
            break  # Setelah menghapus sufiks, berhenti mencari sufiks lainnya

    return word


def stemming(tokens, kata_dasar):
    """
    Fungsi untuk melakukan stemming pada daftar token dengan membandingkan kata dengan kata dasar.
    """
    changes = []  # Daftar untuk menyimpan perubahan kata
    stemmed_tokens = []
    for word in tokens:
        # Pastikan token adalah string
        if isinstance(word, str):
            stemmed_word = simple_stemming(word, kata_dasar)
        else:
            stemmed_word = word  # Jika bukan string, biarkan tanpa perubahan
        stemmed_tokens.append(stemmed_word)
        if word != stemmed_word:  # Catat perubahan kata
            changes.append((word, stemmed_word))

    # Cetak perubahan kata jika ada
    if changes:
        print("Kata yang diganti:", changes)

    return stemmed_tokens

# Memuat daftar kata dasar dari file 'kata-dasar-all.txt'
kata_dasar = load_kata_dasar("kata-dasar-all.txt")

# Aplikasikan fungsi stemming pada data
data['stemming'] = data['stopwords'].apply(lambda tokens: stemming(tokens, kata_dasar))

# Tampilkan data dengan kolom 'stemming'
data

```
